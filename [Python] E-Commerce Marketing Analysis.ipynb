<strong><font size="4"> Part 1: What is the highest share of orders/transactions? </strong></font>
<br><br><strong><font size="3"> Transaction Share by Country </strong></font>


```python
# Group by country and count the unique transaction_id
country_orders = df.groupby('country')['transaction_id'].nunique().reset_index()

# Create new variable transaction_share = transaction_id / sum transaction id 
country_orders['transaction_share'] = country_orders['transaction_id'] / country_orders['transaction_id'].sum()
```

```python
# Displaying the Countries share of transactions
country_with_highest_share = country_orders.sort_values('transaction_share').head(4)

# Save the DataFrame to an Excel file
country_orders.to_excel('country_orders.xlsx', index=False)

print(country_orders)

```

```python
# Sort countries by transaction count
top_countries = country_orders.sort_values('transaction_share', ascending=False).head(10)

# Set up the plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='transaction_id', y='country', data=top_countries, palette='viridis')

# Add titles and labels
plt.title('Top 4 Countries by Amount of Transactions', fontsize=16)
plt.xlabel('Transactions', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.xlim(0, top_countries['transaction_id'].max() * 1.1)  # Extend x-axis limit for better visualization

# Add value labels centered on the bars
for index, value in enumerate(top_countries['transaction_id']):
    ax.text(value, index, f'{value:.0f}', color='black', ha='left', va='center', fontsize=10)

# Save the plot as an image
plt.savefig('countries_transactions.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.tight_layout()
plt.show()
```

<strong><font size="4"> Part 2: Conversion Rate(CR) & Average Daily Revenue </strong></font>

<strong><font size="3"> Conversion Rate by Channels </strong></font>

```python
# Group by channel_id to calculate the number of unique sessions and transactions
transactions = traffic.groupby('channel_id').agg(unique_sessions=('session_id', 'nunique'),
                                                 unique_transactions=('transaction_id', 'nunique')
                                                ).reset_index()

# Calculate Conversion Rate (CR)
transactions['conversion_rate'] = transactions['unique_transactions'] / transactions['unique_sessions']
print(transactions[['channel_id','unique_transactions','unique_sessions', 'conversion_rate']])

# Save the DataFrame to an Excel file
transactions.to_excel('channel_transactions.xlsx', index=False)
```

```python
# Set up the figure and axes
fig, ax1 = plt.subplots(figsize=(20,6))

# Plot bar chart for unique_sessions and unique_transactions
ax1.bar(transactions['channel_id'] - 0.2, transactions['unique_sessions'], width=0.4, label='Unique Sessions', color='lightblue')
ax1.bar(transactions['channel_id'] + 0.2, transactions['unique_transactions'], width=0.4, label='Unique Transactions', color='orange')
ax1.set_xlabel('Channel ID', fontsize=12)
ax1.set_ylabel('Count', fontsize=12)
ax1.legend(loc='upper left')

# Create a second y-axis to plot conversion_rate
ax2 = ax1.twinx()
ax2.plot(transactions['channel_id'], transactions['conversion_rate'], color='red', marker='o', label='Conversion Rate', linewidth=2)
ax2.set_ylabel('Conversion Rate', fontsize=12)

# Add title, grid, and legend
plt.title('Channel Performance: Sessions, Transactions, and Conversion Rate', fontsize=16)
ax1.grid(True, which='both', axis='both', linestyle='--', linewidth=0.5)
ax2.legend(loc='upper right')


# Save the plot as an image (useful for embedding in reports)
plt.savefig('channel_performance.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

<strong> Insights: </strong>
- Channel 1, 2, & 11 bring the less traffic, however their conversion rates are the highest.
- Channel 1 has the highest conversion rate, it is really efficient in driving conversions. It is important to consider that it has only 1493 users.
- Channel 2 has a solid conversion rate of approx. 4% with a large number of sessions (41985). This channel combines a large volume of sessions with good efficiency in conversion rates.
- Channel 11 has a 3.7% of conversion rate with only 6897 sessions.
- Our top 4 under performing channels are Channel 4, 10, 5, & 9. 
- Definitely we should invest in Channels 1,2, & 11.
- We should investigate and rethink resource allocation for under performing channels.


```python
# Set up the plot
plt.figure(figsize=(20, 6))

# Define ax to capture the axes object from sns.barplot
ax = sns.barplot(x='channel_id', y='conversion_rate', data=transactions, palette='viridis')

# Adding titles and labels
plt.title('Channel ID', fontsize=16)
plt.xlabel('Conversion Rates', fontsize=12)
plt.ylabel('Conversion Rates by Channel', fontsize=12)

# Adding Value Labels
for index, value in enumerate(transactions['conversion_rate']):
    # Getting the x-coordinate of the center of the bar
    x_position = index
    # Adding the text label centered in the bar, slightly above the bar
    ax.text(x_position, value, f'{value:.2f}', ha='center', va='bottom', fontsize=10, color='black')

# Save the plot as an image
plt.savefig('channel_performance_2.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.tight_layout()
plt.show()
```

<strong><font size="3"> Average Daily Revenue per Channel </strong></font>

```python
# Merge Google Analytics data with orders refunds to get revenue per channel
revenue_data = pd.merge(traffic, refund, left_on='transaction_id', right_on='order_Id')

# Group by channel_id and aggregate total revenue and the number of days
revenue_per_channel = revenue_data.groupby('channel_id').agg(total_revenue=('final_revenue', 'sum'),
                                                         days_active=('order_date', 'nunique')  
                                                         ).reset_index()

# Calculate average daily revenue
revenue_per_channel['avg_daily_revenue'] = (revenue_per_channel['total_revenue'] / revenue_per_channel['days_active']).round()
print(revenue_per_channel[['channel_id','avg_daily_revenue']])

# Save the DataFrame to an Excel file
revenue_per_channel.to_excel('revenue_per_channel.xlsx', index=False)
```

```python
# Set up the plot
plt.figure(figsize=(20, 6))

# Define ax to capture the axes object from sns.barplot
ax = sns.barplot(x='channel_id', y='avg_daily_revenue', data=revenue_per_channel, palette='viridis')

# Adding titles and labels
plt.title('Channel ID', fontsize=16)
plt.xlabel('Channels', fontsize=12)
plt.ylabel('Average Daily Revenue (€)', fontsize=12)

# Adding Value Labels
for index, value in enumerate(revenue_per_channel['avg_daily_revenue']):
    # Getting the x-coordinate of the center of the bar
    x_position = index
    # Adding the text label centered in the bar, slightly above the bar
    ax.text(x_position, value, f'{value:.2f}', ha='center', va='bottom', fontsize=10, color='black')

# Save the plot as an image
plt.savefig('average_daily_revenue.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.tight_layout()
plt.show()
```

<strong> Insights: </strong>
- Channel 7 brings an average daily revenue of 351524 Euro. We can consider as our revenue driver.
- Channel 2, 3, 8 & 13 bring an average daily revenue of 135987 Euro. We should consider investigating why this channels are performing well and if it possible to optimize them in order to reach Channel 7 level.
- We should focus our marketing strategy on Channels 7, 2, 3, 8, & 13.
- We should evaluate channels 1, 4, and 10.


 <br><strong><font size="4">Part 3: Average Refund Rates</strong></font>
 <br><br><strong><font size="3"> Refund Rates by Channel </strong></font>

```python
# Calculate total orders and refunded orders
refund_data = df.copy()
refund_data['is_refunded'] = refund_data['refunded_revenue'] > 0

# Group by channel_id or overall and calculate refund rates
refund_rate = refund_data.groupby('channel_id').agg(total_orders=('order_Id', 'nunique'),
                                                    refunded_orders=('is_refunded', 'sum')
                                                    ).reset_index()

# Calculate refund rate
refund_rate['refund_rate'] = refund_rate['refunded_orders'] / refund_rate['total_orders']
print(refund_rate[['channel_id', 'refund_rate']])

# Save the DataFrame to an Excel file
refund_rate.to_excel('refund_rate.xlsx', index=False)
```

```python
# Set up the plot
plt.figure(figsize=(20, 6))

# Define ax to capture the axes object from sns.barplot
ax = sns.barplot(x='channel_id', y='refund_rate', data=refund_rate, palette='viridis')

# Adding titles and labels
plt.title('Refund Rate by Channels', fontsize=16)
plt.xlabel('Channels', fontsize=12)
plt.ylabel('Refund Rate', fontsize=12)

# Adding Value Labels
for index, value in enumerate(refund_rate['refund_rate']):
    # Getting the x-coordinate of the center of the bar
    x_position = index
    # Adding the text label centered in the bar, slightly above the bar
    ax.text(x_position, value, f'{value:.2f}', ha='center', va='bottom', fontsize=10, color='black')

# Save the plot as an image
plt.savefig('refund_rate_by_channels.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.tight_layout()
plt.show()
```

<strong> Insights: </strong>
- Try to understand Channels 1, 10, 11, & 13. This channels have the higher refund rates. It would be interesting to analyze customer feedback, review product quality, and evaluate the user process to identify potential pain points.
- Channels 2 & 3 are performing well. Analyzing their success can give good insights that can be applied over the rest of the channels.
- Refunds provide a way to understand customer satisfaction and how E-commerce can improve in their offerings.
- Understanding refund causes will lead to improve customer experience.


<strong><font size="3"> Refund Rates by Country </strong></font>

<strong><font size="3"> Refund Rates by Date </strong></font>

```python
# Calculate total orders and refunded orders
refund_data_day = df.copy()
refund_data_day['is_refunded'] = refund_data_day['refunded_revenue'] > 0

# Group by order_date and calculate total and refunded orders
refund_rate_day = refund_data_day.groupby('order_date').agg(
    total_orders=('order_Id', 'nunique'),
    refunded_orders=('is_refunded', 'sum')
).reset_index()

# Calculate refund rate
refund_rate_day['refund_rate'] = refund_rate_day['refunded_orders'] / refund_rate_day['total_orders']

# Display the result
print(refund_rate_day[['order_date', 'refund_rate']])
```

```python
# Set up the plot
plt.figure(figsize=(12, 6))

# Define ax to capture the axes object from sns.barplot
ax = sns.barplot(x='order_date', y='refund_rate', data=refund_rate_day, palette='viridis')

# Adding titles and labels
plt.title('Refund Rate by Day', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Refund Rate', fontsize=12)

# Adding Value Labels
for index, value in enumerate(refund_rate_day['refund_rate']):
    # Getting the x-coordinate of the center of the bar
    x_position = index
    # Adding the text label centered in the bar, slightly above the bar
    ax.text(x_position, value, f'{value:.2f}', ha='center', va='bottom', fontsize=10, color='black')

# Save the plot as an image
plt.savefig('refund_rate_by_date.png', dpi=300, bbox_inches='tight')       
    
# Show the plot
plt.tight_layout()
plt.show()
```

<strong><font size="3"> Part 4: Efficiency of Marketing Channels </strong></font>

```python
# Grouping by 'channel_id' and aggregating the 'revenue' column as sum
revenue_grouped = df.groupby('channel_id').agg({'final_revenue': 'sum',
                                                'refunded_revenue': 'sum'
                                                }).reset_index()

# Add a new column 'total_revenue'
revenue_grouped['total_revenue'] = (revenue_grouped['final_revenue'])

# Calculate the daily average net revenue
daily_average_revenue = revenue_grouped['final_revenue'] / 3

# Estimate total revenue for 30 days
revenue_grouped['total_revenue'] = daily_average_revenue * 30

# Add a new column 'mktg_spend' with the specified values
revenue_grouped['mktg_spend'] = [830, 172810, 12681, 14250, 258171, 75190, 218305, 25421,63146]

# Add a new column 'ROAS' Return on Advertisements
revenue_grouped['ROAS'] = revenue_grouped['total_revenue']/revenue_grouped['mktg_spend']

# Save the DataFrame to an Excel file
revenue_grouped.to_excel('revenue_grouped.xlsx', index=False)

# Display the first 50 rows of the aggregated table
print(revenue_grouped.head(50))
```

```python
# Set up the plot
plt.figure(figsize=(20, 6))

# Define ax to capture the axes object from sns.barplot
ax = sns.barplot(x='channel_id', y='ROAS', data=revenue_grouped, palette='viridis')

# Adding titles and labels
plt.title('Return On Ad Spends', fontsize=16)
plt.xlabel('Channels', fontsize=12)
plt.ylabel('ROAS', fontsize=12)

# Adding Value Labels
for index, value in enumerate(revenue_grouped['ROAS']):
    # Getting the x-coordinate of the center of the bar
    x_position = index
    # Adding the text label centered in the bar, slightly above the bar
    ax.text(x_position, value, f'{value:.2f}', ha='center', va='bottom', fontsize=10, color='black')

# Save the plot as an image
plt.savefig('return_on_advertisement.png', dpi=300, bbox_inches='tight') 

# Show the plot
plt.tight_layout()
plt.show()
```
